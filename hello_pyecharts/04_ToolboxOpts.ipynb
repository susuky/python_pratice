```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker

c = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("商家A", Faker.values())
    .add_yaxis("商家B", Faker.values())
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Bar-显示 ToolBox"),
        toolbox_opts=opts.ToolboxOpts(is_show=True),
        legend_opts=opts.LegendOpts(is_show=False),
    )
)
c.render_notebook()
```

# 選擇特定toolbox

```python
c.set_global_opts(
        title_opts=opts.TitleOpts(title="Bar-显示 ToolBox"),
        toolbox_opts=opts.ToolboxOpts(
            is_show=True,
            feature=opts.ToolBoxFeatureOpts(
                save_as_image=None,
                restore=None,
                data_view=None,
                data_zoom=None,
                magic_type=None,
                brush=None,
            )
        ),
        legend_opts=opts.LegendOpts(is_show=False),
    )

c.render_notebook()
```

# 選取

```python
c.set_global_opts(
        title_opts=opts.TitleOpts(title="Bar-显示 ToolBox"),
        toolbox_opts=opts.ToolboxOpts(
            is_show=True,
            feature=opts.ToolBoxFeatureOpts(
                save_as_image=opts.ToolBoxFeatureSaveAsImageOpts(
                    title='Download plot as png',
                    type_='png',
                    background_color='white',
                    pixel_ratio=1,
                ),
                restore=opts.ToolBoxFeatureRestoreOpts(
                    title='Restore',
                ),
                data_view=opts.ToolBoxFeatureDataViewOpts(
                    title='View data',
                    is_read_only=False,
                    lang=['Data', 'Close', 'Refresh'],
                    background_color='#272727',
                    text_area_color='#272727',
                    text_area_border_color='#F0F0F0',
                    text_color='#F0F0F0',
                    button_color='#FF7575',
                    button_text_color="#000000",
                ),
                data_zoom=opts.ToolBoxFeatureDataZoomOpts(
                    is_show=True,
                    zoom_title='Zoom',
                    back_title='Zoom restore',
                    xaxis_index=[0],  # 選擇能控制的xaxis
                    yaxis_index=[0],  # 選擇能控制的yaxis
                ),
                magic_type=opts.ToolBoxFeatureMagicTypeOpts(
                    # 文字顯示有bug
                    is_show=True,
                    #tiled_title='tiled',
                    line_title='Line',
                    bar_title='Bar',
                    stack_title='Stack',
                ),
                brush=opts.ToolBoxFeatureBrushOpts(
                    rect_title='Box Select',
                    polygon_title='Polygon Select',
                    line_x_title='H Select',
                    line_y_title='V Select',
                    keep_title='Keep Select',
                    clear_title='Clear Select',                    
                ), # brush要另外設定BrushOpts才會有效
            )
        ),
        brush_opts=opts.BrushOpts( 
            # 改icon要去ToolBoxFeatureBrushOpts改
            ["rect", "polygon", "lineX", "lineY", "keep", "clear"],
            brush_mode='multiple',
            brush_style=None
        ),
        legend_opts=opts.LegendOpts(is_show=False),
    )

c.render_notebook()
```

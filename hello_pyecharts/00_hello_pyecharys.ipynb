# ThemeType

````python
from pyecharts import options as opts
from pyecharts.charts import Bar, Page
from pyecharts.faker import Collector, Faker
from pyecharts.globals import ThemeType


def theme_default(theme: ThemeType.LIGHT) -> Bar:
    c = (
        Bar(init_opts=opts.InitOpts(theme=theme))
        .add_xaxis(Faker.choose())
        .add_yaxis("商家A", Faker.values())
        .add_yaxis("商家B", Faker.values())
        .add_yaxis("商家C", Faker.values())
        .add_yaxis("商家D", Faker.values())
        .set_global_opts(title_opts=opts.TitleOpts(theme))
    )
    return c

Page().add(*(theme_default(getattr(ThemeType, themetype))
             for themetype in dir(ThemeType) 
             if not themetype.startswith('__') and themetype != 'BUILTIN_THEMES'
            )
          ).render_notebook()
````

## 輸入資料須自行轉換成以下格式
Python        |  JSON
------------ |   --------------
int, float   |   number
str           |  string
bool           | boolean
dict            |object
list            |array

**numpy**, **pands**都不能直接輸入，**tuple**也不行

```python
[int(i) for i in something]
[float(i) for i in something]
[str(i) for i in something]

Series.tolist()
Numpy.tolist()
DataFrame.values.tolist()
```

```python
c = (
    Kline()
    .add_xaxis([f"2017/7/{i+1}" for i in range(31)])
    .add_yaxis("kline", data)
)
c.render_notebook()
```

```python
c.set_global_opts(
        xaxis_opts=opts.AxisOpts(is_scale=True),
        yaxis_opts=opts.AxisOpts(
            is_scale=True,
            splitarea_opts=opts.SplitAreaOpts( # yaxis分隔區
                is_show=True, 
                areastyle_opts=opts.AreaStyleOpts(opacity=1)
            ), 
        ),
    
        title_opts=opts.TitleOpts(title="Kline-basic"), # 標題
        datazoom_opts=[
            opts.DataZoomOpts(type_="inside"), # 滑鼠拉動、滾動
            opts.DataZoomOpts(type_="slider"), # 滑動條
        ], 
    )
c.render_notebook()
```

# itemstyle

```python
c = (
    Kline()
    .add_xaxis([f"2017/7/{i+1}" for i in range(31)])
    .add_yaxis(
        "kline", 
        data,
        itemstyle_opts=opts.ItemStyleOpts( # 可以選顏色
            color="#ec0000",  # 陽
            color0="#00da3c", # 陰
            border_color="#8A0000",
            border_color0="#008F28",
        ),
    )
)

c.set_global_opts(
        xaxis_opts=opts.AxisOpts(is_scale=True),
        yaxis_opts=opts.AxisOpts(
            is_scale=True,
            splitarea_opts=opts.SplitAreaOpts(
                is_show=True, 
                areastyle_opts=opts.AreaStyleOpts(opacity=1, color=['#3C3C3C', '#7B7B7B']) # 可以選雙色
            ), 
            splitline_opts=opts.SplitLineOpts( 
                is_show=True, linestyle_opts=opts.LineStyleOpts(width=1)
            ),
        ),
    
        title_opts=opts.TitleOpts(title="Kline-basic"), 
        datazoom_opts=[
            opts.DataZoomOpts(type_="inside"), 
            opts.DataZoomOpts(type_="slider", pos_bottom="-2%"), 
        ], 
)
c.render_notebook()
```

```python
from pyecharts.globals import ThemeType
```

```python
c = (
    Kline(
        init_opts=opts.InitOpts(
            theme=ThemeType.CHALK, 
            bg_color="#23262F"
        )
    )
    .add_xaxis([f"2017/7/{i+1}" for i in range(31)])
    .add_yaxis(
        "kline", 
        data,
        itemstyle_opts=opts.ItemStyleOpts(
            color="#FF0000",  # 陽
            color0="#28FF28", # 陰
            border_color="#FF0000",
            border_color0="#28FF28",
        ),
    )
)

c.set_global_opts(
        xaxis_opts=opts.AxisOpts(is_scale=True),
        yaxis_opts=opts.AxisOpts(
            is_scale=True,
            splitline_opts=opts.SplitLineOpts( 
                is_show=True, linestyle_opts=opts.LineStyleOpts(
                    width=1, color='#7B7B7B', opacity=0.5
                )
            ),
        ),
    
        title_opts=opts.TitleOpts(title="Kline-basic"), 
        datazoom_opts=[
            opts.DataZoomOpts(type_="inside"), 
            opts.DataZoomOpts(type_="slider", pos_bottom="-2%"), 
        ], 
)
c.render_notebook()
```

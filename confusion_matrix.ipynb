```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
preds = np.random.randint(0, 5, size=10000)
ys = np.concatenate([np.random.randint(0, 5, size=2000), preds[2000:]])
```

```python
def confusion_matrix(preds, ys):
    nunique = len(np.unique(ys))
    cm = np.zeros((nunique, nunique), dtype=int)
    for i, j in zip(ys, preds):
        cm[i, j] += 1
    return cm

confusion_matrix(preds, ys)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
formats = ((None, 'd'), ('true', '.2%'))
for fmt in formats:
    cm = confusion_matrix(ys, preds, normalize=fmt[0])
    sns.heatmap(cm, annot=True, fmt=fmt[1], cmap='Blues')
    plt.ylabel('True')
    plt.xlabel('Predict')    
    plt.show()
```
